```python
import yfinance as yf
import matplotlib.pyplot as plt

# 获取股票指标数据
def get_stock_indicators(symbol):
    stock = yf.Ticker(symbol)
    key_statistics = stock.info
    return key_statistics

# 获取股票平均净资产收益率、市盈率、经营现金流和净利润数据
def get_indicators_data(symbol):
    indicators_data = get_stock_indicators(symbol)
    roa = indicators_data.get('returnOnAssets', None)
    pe_ratio = indicators_data.get('forwardPE', None)
    operating_cashflow = indicators_data.get('operatingCashflow', None)
    net_profit = indicators_data.get('netIncome', None)
    return roa, pe_ratio, operating_cashflow, net_profit

# 可视化四个指标数据
def visualize_indicators(symbol, roa, pe_ratio, operating_cashflow, net_profit):
    labels = ['ROA', 'PE Ratio', 'Operating Cashflow', 'Net Profit']
    values = [roa, pe_ratio, operating_cashflow, net_profit]

    plt.bar(labels, values, color=['blue', 'green', 'orange', 'red'])
    plt.title(f'{symbol} Indicators')
    plt.ylabel('Value')
    plt.xticks(rotation=45)
    plt.show()

# 示例用法
if __name__ == "__main__":
    # 股票代码
    symbol = 'AAPL'

    # 获取股票指标数据
    roa, pe_ratio, operating_cashflow, net_profit = get_indicators_data(symbol)

    # 可视化四个指标数据
    visualize_indicators(symbol, roa, pe_ratio, operating_cashflow, net_profit)
```
